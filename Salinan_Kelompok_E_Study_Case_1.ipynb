```python
import pandas as pd

# Step 1: Ensure all values in 'amenities' are strings to avoid errors with .str accessor
listings['amenities'] = listings['amenities'].astype(str).fillna("")

# Step 2: Clean the amenities column and split items into lists
listings['amenities'] = listings['amenities'].str.strip("[]").str.replace("'", "").str.replace('"', "").str.split(", ")

# Step 3: Ensure all entries in standardized_amenities are lowercase for uniformity
standardized_amenities = {
    "hdtv": [
        "120 hdtv with amazon prime video", "120 hdtv with apple tv", "24 hdtv with chromecast", "32 hdtv",
        "32 hdtv with amazon prime video", "32 hdtv with chromecast", "32 hdtv with hbo max",
        "32 hdtv with netflix", "32 hdtv with roku", "37 tv with amazon prime video", "40 hdtv with netflix",
        "40 hdtv with roku", "42 hdtv with amazon prime video", "42 hdtv with netflix", "43 hdtv with netflix",
        "45 hdtv with netflix", "46 hdtv with premium cable", "48 hdtv with amazon prime video",
        "48 hdtv with apple tv", "48 hdtv with premium cable", "50 hdtv", "50 hdtv with amazon prime video",
        "50 hdtv with chromecast", "50 hdtv with netflix", "50 hdtv with standard cable", "51 hdtv",
        "55 hdtv with amazon prime video", "55 hdtv with netflix", "55 hdtv with roku", "60 hdtv with netflix",
        "65 hdtv", "65 hdtv with amazon prime video", "65 hdtv with netflix", "75 hdtv with netflix"
    ],
    "refrigerator": [
        "amana refrigerator", "americana refrigerator", "blomberg refrigerator", "bosch refrigerator",
        "electrolux refrigerator", "frigidaire refrigerator", "fridgedaire refrigerator",
        "ge profile refrigerator", "ge refrigerator", "inc refrigerator", "kenmore refrigerator",
        "lg refrigerator", "lg smart tech refrigerator", "miele refrigerator", "samsung refrigerator",
        "sharp refrigerator", "subzero refrigerator", "undercounter refrigerator refrigerator", "whirlpool refrigerator"
    ],
    "sound_system": [
        "alexa sound system with bluetooth and aux", "bose & sony sound system with bluetooth and aux",
        "bose soundbar bluetooth sound system", "bang & olufson sound system", "bluetooth sound system",
        "bose sound system", "bose bluetooth sound system", "bose sound system with bluetooth and aux",
        "google home bluetooth sound system", "sonos bluetooth sound system"
    ],
    "body_soap": [
        "dove unscented body soap", "dove", "dove deep moisture body soap", "dial", "for sensitive skin body soap",
        "irish spring", "neutrogena body wash body soap", "nubian heritage black soap body soap"
    ],
    "wifi": [
        "wifi 100 mbps", "wifi 1000 mbps", "pocket wifi"
    ],
    # Continue for other amenities as before, converted to lowercase
}

# Step 4: Function to standardize amenities based on the dictionary
def standardize_amenities(amenities_list):
    standardized_list = []
    for amenity in amenities_list:
        found = False
        amenity_lower = amenity.lower()
        for key, synonyms in standardized_amenities.items():
            if amenity_lower in synonyms:
                standardized_list.append(key)
                found = True
                break
        if not found:
            standardized_list.append(amenity)  # keep original if no synonym found
    return list(set(standardized_list))  # remove duplicates

# Apply standardization
listings['amenities'] = listings['amenities'].apply(standardize_amenities)

# Step 5: Use pd.get_dummies to create binary columns for each unique standardized amenity
amenities_dummies = listings['amenities'].str.join('|').str.get_dummies()

# Step 6: Join the dummies DataFrame back to the original DataFrame
listings_normalized = pd.concat([listings, amenities_dummies], axis=1)

# Step 7: Drop the original amenities column if desired
listings_normalized = listings_normalized.drop(columns=['amenities'])

# Display the resulting DataFrame
listings_normalized.head()
```

```python
# Example 8.2 Logit Model of Economic Education
# Binomial Regression Model: Probit and Logit
# Logit Model using statsmodels package

import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

data = pd.read_csv('http://web.pdx.edu/~crkl/ceR/data/grade.txt',sep='\s+', nrows=32)
data = data.dropna()
print(data.describe())

X = data[['GPA','TUCE','PSI']]
y = data['GRADE']
model1 = sm.Logit(y,sm.add_constant(X)).fit()
model1.summary()

model2 = sm.Logit.from_formula('GRADE ~ GPA + TUCE + PSI',data=data).fit()
model2.summary()

# same as model2
model3 = smf.logit('GRADE ~ GPA + TUCE + PSI', data=data).fit()
model3.summary()

# using GLM
model4 = smf.glm(formula='GRADE ~ GPA + TUCE + PSI',
                family=sm.families.Binomial(),
                data=data).fit()
model4.summary()

# probit model interpretation
model2.params
# predicted probability of each obs
model2.predict(data)
# marginal effects at the mean of each regressor
margeff = model2.get_margeff('mean')
margeff.summary()
```
